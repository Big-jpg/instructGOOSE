```python
from transformers import AutoTokenizer, AutoModelForCausalLM

from instruct_goose.env import TextEnv
from instruct_goose.utils import load_yaml
```

```python
config = load_yaml("../configs/sentiment_config.yml")
```

```python
tokenizer = AutoTokenizer.from_pretrained(config["model"]["tokenizer_path"])
```

```python
model = AutoModelForCausalLM.from_pretrained(config["model"]["model_path"])
```

```python
prompt = "Tell me about your day"
```

```python
prompt_ids = tokenizer(prompt, return_tensors="pt")
```

```python
prompt_ids["input_ids"][0]
```

```python
env = TextEnv(
    model, tokenizer, observation_input=prompt_ids["input_ids"][0], context_length=1024
)
```

```python
env.reset()
```

```python
len(env.input_token_ids)
```

```python
env.action_space
```
